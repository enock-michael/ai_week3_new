```python
import spacy
from spacy import displacy

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Sample Amazon product reviews
reviews = [
    "I love the new Apple iPhone 13! The camera quality is incredible.",
    "The Samsung Galaxy S21 is just okay. The battery life could be better.",
    "I recently purchased the Bose QuietComfort 35 II headphones, and I am very satisfied with their noise cancellation.",
    "The Dell XPS 13 laptop is excellent for work and play. Highly recommend!",
    "I had a terrible experience with the Sony WH-1000XM4 headphones. They broke within a week."
]

# Function to perform NER and sentiment analysis
def analyze_reviews(reviews):
    extracted_entities = []
    sentiments = []
    
    for review in reviews:
        doc = nlp(review)
        
        # Extract entities
        entities = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in ["PRODUCT", "ORG"]]
        extracted_entities.append(entities)
        
        # Simple rule-based sentiment analysis
        if "love" in review or "excellent" in review or "incredible" in review or "highly recommend" in review:
            sentiments.append("Positive")
        elif "terrible" in review or "okay" in review or "could be better" in review:
            sentiments.append("Negative")
        else:
            sentiments.append("Neutral")

    return extracted_entities, sentiments

# Analyze the reviews
entities, sentiments = analyze_reviews(reviews)

# Display the results
for i, review in enumerate(reviews):
    print(f"Review: {review}")
    print(f"Extracted Entities: {entities[i]}")
    print(f"Sentiment: {sentiments[i]}\n")
```
